```python
#Grading system
marks= int(input("Enter your marks:"))
if(marks>=90 and marks<100):
    print("YOU GOT: PASS AND Grade A")
elif(marks>=80 and marks<90):
    print("YOU GOT: PASS AND Grade B")
elif(marks>=60 and marks<80):
    print("YOU GOT: PASS AND Grade C")
elif(marks>=48 and marks<60):
    print("YOU GOT: PASS AND Grade D")
elif(marks>-47 and marks>0):
    print("YOU GOT: FAIL")
else:
    print("ENTER THE CORRECT MARKS!! LIEING BETWEEN 0 TO 100.")
```

```python
#SIMPLE CALCULATOR
num1=int(input("ENTER THE FIRST NUMBER:"))
num2=int(input("ENTER THE SECOND NUMBER:"))
op=input("ENTER THE OPERATOR FROM{+,-,*,/,%,**} :")
if(op=="+"):
   print("THE SUM IS:",num1+num2)
elif(op=="-"):
   if(num1<num2):
      print("THE DIFFERENCE IS:",num2-num1)
   else:
      print("THE DIFFERENCE IS:",num1-num2)
elif(op=="*"):
   print("THE PRODUCT IS:",num1*num2)
elif(op=="/"):
   print("THE QUOTIENT IS:",num1/num2)
elif(op=="%"):
   print("THE REMAINDER IS:",num1%num2)
elif(op=="**"):
   print("THE POWER IS:",num1**num2)
else:
   print("ENTER THE CORRECT OPERATOR!!")
```
