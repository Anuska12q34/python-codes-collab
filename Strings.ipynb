```python
#Grading system
marks= int(input("Enter your marks:"))
if(marks>=90 and marks<100):
    print("YOU GOT: PASS AND Grade A")
elif(marks>=80 and marks<90):
    print("YOU GOT: PASS AND Grade B")
elif(marks>=60 and marks<80):
    print("YOU GOT: PASS AND Grade C")
elif(marks>=48 and marks<60):
    print("YOU GOT: PASS AND Grade D")
elif(marks>-47 and marks>0):
    print("YOU GOT: FAIL")
else:
    print("ENTER THE CORRECT MARKS!! LIEING BETWEEN 0 TO 100.")
```
