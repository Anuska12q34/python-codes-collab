```python
#Aritheatic operators
a=50
b=20

print("sum :", a+b)
print("Diff :",a-b)
print("Mul :", a*b)
print("Div :", a/b)
print("Modular:", a%b)#remainder
print("Power :", a**b)#a^b=a**b
```

```python
#relational operators
a = 50
b = 20

print("a==b:", a==b)
print("a!=b:", a!=b)
print("a>=b:", a>=b)
print("a>b:", a>b)
print("a<=b:", a<=b)
print("a<b:", a<b)
```
